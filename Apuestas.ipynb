```python
equipo_local=Aa.HomeTeam
equipo_visitante=Aa.AwayTeam
marcador_local=Aa.FTHG
marcador_visitante=Aa.FTAG
momio_local=Aa.B365H
momio_visitante=Aa.B365A
Puntos=pd.DataFrame({'puntos':1000*np.ones(Aa.HomeTeam.unique().size)}, index=Aa.HomeTeam.unique())
n_media_movil=2*Aa.HomeTeam.unique().size
media_movil=[]


aument_predice=[]
aument_acierta=[]
aument_apuesta=[]
no_partido_apuesta=[]
gana_apuesta=[]
no_partido=[]
momios_guardados=[]

####################################################
# Criterios para realizar la apuesta

criterio=8  # diferencia entre equipos
momio_minimo=2
PJ=0       # N√∫mero de partidos jugados
EM_min=0.6     # valor m√≠nimo de la media m√≥vil
####################################################

for i in range(equipo_local.size):

  name_local=equipo_local[i]
  name_visitante=equipo_visitante[i]
  resultado=marcador_local[i]-marcador_visitante[i]

  winner_real='bla-bla'

  if resultado>0:
    winner_real=name_local
    
  elif resultado<0:
    winner_real=name_visitante
    

  if (Puntos.loc[name_local]>Puntos.loc[name_visitante]).bool():
    winner_esperado=name_local
    momio=momio_local[i]
  else:
    winner_esperado=name_visitante
    momio=momio_visitante[i]

  Predecir=np.abs(Puntos.loc[name_local]-Puntos.loc[name_visitante])>criterio
  Predecir=Predecir.bool()

  if Predecir:
    aument_predice.append(1)
    no_partido.append(i)
    aument_acierta.append(winner_real==winner_esperado)
    media_movil.append(sum(aument_acierta[-n_media_movil:])/n_media_movil)

  Apostar=Predecir and i>PJ and media_movil[-1]>=EM_min and momio>momio_minimo

  if Apostar:
    aument_apuesta.append(1)
    no_partido_apuesta.append(i)
    gana_apuesta.append(winner_real==winner_esperado)
    if winner_real==winner_esperado:
      momios_guardados.append(momio)
  

  
  ########################################################################
  # Bloque borrable, no altera el c√≥digo
  #######################################################
  if i==0:
    Cuenta_init=250
    Cuenta=Cuenta_init
  Monto_por_apuesta=0.1*Cuenta_init
  if Apostar:
    Cuenta+=-Monto_por_apuesta
    if winner_real==winner_esperado:
      Cuenta+=Monto_por_apuesta*momio
  if Cuenta<0:
    print('------------------------------Cuenta rota-----------------------------')
      
  #######################################################
  # Termina el bloque borrable
  ########################################################################
    

  if resultado>=0:  # N1: ganador, N2:perdedor
    N1=name_local
    N2=name_visitante
  else:
    N1=name_visitante
    N2=name_local
  if resultado==0:
    r=0.5
  else:
    r=1


  Ra=Puntos.loc[N1]
  Rb=Puntos.loc[N2]

  Ves_a=1/(1+10**((Rb-Ra)/400))   # Ves: valor esperado
  Ves_b=1/(1+10**((Ra-Rb)/400))

  Vob_a=r
  Vob_b=1-r

  Sa=32*(Vob_a-Ves_a)
  Sb=32*(Vob_b-Ves_b)

  Puntos.loc[N1]+=Sa
  Puntos.loc[N2]+=Sb

a_acierta=np.array(aument_acierta)
a_predice=np.array(aument_predice)
gana_apuesta=np.array(gana_apuesta)
apuestas=np.array(aument_apuesta)
#no_partido=np.array(no_partido)
#media_movil=np.array(media_movil)

acierta_acumulado=np.add.accumulate(a_acierta)
gana_apuesta_acumulado=np.add.accumulate(gana_apuesta)
predice_acumulado=np.add.accumulate(a_predice)
apuestas_acumuladas=np.add.accumulate(apuestas)

plt.plot(no_partido, media_movil, no_partido, EM_min*np.ones(len(no_partido)))
plt.title(str(Aa.Date.unique()[[0,-1]]))

if len(apuestas_acumuladas)>0:
  print('Apuestas acertadas: ', gana_apuesta_acumulado[-1])
  print('Apuestas hechas: ', apuestas_acumuladas[-1])
  promedio=gana_apuesta_acumulado[-1]/apuestas_acumuladas[-1]
  print('Promedio de apuestas ganadas: ', promedio)
  print('momio medio: ', np.mean(momios_guardados))
  print('Esperanza matem√°tica: ', np.mean(momios_guardados)*promedio-1)
  print('Valor inicial de la cuenta: ', Cuenta_init)
  print('Monto por apuesta: ', Monto_por_apuesta)
  print('Valor final de la cuenta: ', Cuenta)
else:
  print('No se realizaron apuestas')
```
